```python
import pandas as pd
adult_training = pd.read_csv("https://raw.githubusercontent.com/bforoura/IDS/main/HW4/adult_training.csv")
```

```python
# Workclass and sex

p_crosstab = pd.crosstab(adult_training.workclass, adult_training.sex)

print(p_crosstab)
```

```python
p_crosstab = pd.crosstab(adult_training.sex, adult_training["marital-status"])

print(p_crosstab)
```

```python
adultOver40 = adult_training[adult_training["age"]>40]
p_crosstab = pd.crosstab(adultOver40.sex, adultOver40["marital-status"])

print(p_crosstab)

```

```python

from scipy import stats
import matplotlib.pyplot as plt

adult_training['scaled_education'] = stats.zscore(adult_training.education, nan_policy='omit')

adult_training.education.plot.density(color="red")
plt.title('Density plot for Unscaled Education')
plt.show()


adult_training.scaled_education.plot.density(color="blue")
plt.title('Density plot for Scaled Education')
plt.show()

## Identifying Outliers:

adult_training_outliers = adult_training.query('(scaled_education > 3 | scaled_education< -3)')['scaled_education']

adult_training_outliers.plot.density()
plt.title('Density Plot for Scaled Education Outliers')
plt.show()
```
