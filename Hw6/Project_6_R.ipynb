```r
library(readr)
library(rpart)

# Load the data
loan_Training <- read_csv("https://raw.githubusercontent.com/bforoura/IDS/main/HW6/Loans_Training.csv", show_col_types = FALSE)
loan_Test <- read_csv("https://raw.githubusercontent.com/bforoura/IDS/main/HW6/Loans_Test.csv", show_col_types = FALSE)

# Convert Approval to a factor
loan_Training$Approval <- as.factor(loan_Training$Approval)

# Check and modify column names if necessary
# Replace 'Debt-to-Income_Ratio' and 'FICO_Score' with the correct column names from your dataset
cart_model <- rpart(Approval ~ `Debt-to-Income Ratio` + `FICO Score`, data = loan_Training, method = "class")

# Print the model summary
print(cart_model)



# Data pre-processing
# Ensure 'Approval' is a factor
# loans_training$Approval <- as.factor(loans_training$Approval)

# Building the CART model
# Predicting 'Approval' based on 'Debt-to-Income Ratio' and 'FICO Score'
# model <- rpart(Approval ~ Debt-to-Income Ratio + FICO Score, data = loan_Training, method = "class")

# View the model summary
print(cart_model)
```

```r

# Visualize the decision tree and Describe the first few splits in the decision tree.
#The root node of the tree represents the entire dataset.
#The first split is made on the variable that best separates the data on FICO.
#The next nodes represents further splits of the data made on Debt to Income.
#
install.packages("rpart.plot")
library(rpart.plot)
rpart.plot(cart_model, type = 4, extra = 2)
```

```r
# Build a C5.0 model using the same predictors and target as in part 1.
# This time, specify a minimum of 1000 cases per terminal node.
# Visualize the decision tree. Describe the first few splits in the decision tree.
#
library(C50)


# Build C5.0 model
c5_model <- C5.0(Approval ~ `Debt-to-Income Ratio` + `FICO Score`,
                 data = loan_Training,
                 control = C5.0Control(minCases = 1000))

# Visualize the decision tree
plot(c5_model)

# EXTRA CREDIT Please: more detailed text representation
summary(c5_model)

# The root node of the tree represents the entire dataset.
#The first split is made on the variable that best separates the data on FICO.
#The next nodes represents further splits of the data made on Debt to Income.
```
